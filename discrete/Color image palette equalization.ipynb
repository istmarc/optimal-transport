### Color image

**RGB images**

```python
def rgb_images(img):
    # Red
    img_r = img.copy()
    img_r[:, :, (1,2)] = 0.# Green
    img_g = img.copy()
    img_g[:, :, (0,2)] = 0.
    # Blue
    img_b = img.copy()
    img_b[:,:,(0,1)] = 0.
    return (img_r, img_g, img_b)
```

```python
img_r, img_g, img_b = rgb_images(img)
inputimg_r, inputimg_g, inputimg_b = rgb_images(inputimg)
```

```python
def plot_rgb(img_r, img_g, img_b):
    fig, axes = plt.subplots(1, 3, figsize=(10, 2))
    axes[0].imshow(img_r)
    axes[0].set_title("red")
    axes[1].imshow(img_g)
    axes[1].set_title("green")
    axes[2].imshow(img_b)
    axes[2].set_title("blue")
    plt.show()
```

```python
plot_rgb(img_r, img_g, img_b)
```

```python
plot_rgb(inputimg_r, inputimg_g, inputimg_b)
```

**Histogram**

```python
def plot_histogram(img_r, img_g, img_b):
    fig, axes = plt.subplots(1, 3, figsize=(10, 2))
    axes[0].hist(img_r[:, :, 0].flatten(), color = "red")
    axes[0].set_title("red histogram")
    axes[1].hist(img_g[:, :, 1].flatten(), color = "green")
    axes[1].set_title("green histogram")
    axes[2].hist(img_b[:, :, 2].flatten(), color = "blue")
    axes[2].set_title("blue histogram")
    plt.show()
```

```python
plot_histogram(img_r, img_g, img_b)
```

```python
plot_histogram(inputimg_r, inputimg_g, inputimg_b)
```
